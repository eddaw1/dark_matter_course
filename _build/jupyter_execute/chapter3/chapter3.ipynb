# Lecture 3 - Quantum analysis of WIMP nucleus scattering

The hard sphere scattering model yields lots of information, but it isn't perfect. We have assumed that the nucleus and WIMP are hard balls, but we know that on a microscopic level, scattering objects have wave as well as particle properties that will affect the interaction.

What other inputs can we draw on to find out what happens when a WIMP scatters off a nucleus? Let us take an opposite extreme with the modelling and think of the WIMP's wave properties. How do plane waves scatter off an object?

### Classical plane wave scattering

You will recall some optics problems when a plane wave scatters off a slit in a screen. What you see on a screen on the other side is a series of interference fringes - in some places you get zero intensity, and in other places you get an intensity maximum. Interestingly you can show that the interference pattern you get for diffraction off a solid body having the same dimensions as those of the slit in the screen is the same pattern. Anyway, this is nothing like particle behaviour. 

Let us next consider a classical wave problem that has the same look and feel and perhaps a shade more intuition than optical waves diffracting off a slit-shaped solid - plane water waves scattering off a post, as illustrated in the diagram below.

![off_post](./figures/off_post.png)

The incoming wavefronts are vertical planes which you can think of as the crests of the wave moving to the right. There is a post in the water. When the plane waves strike the post, the result can be thought of as a fraction of the plane waves carrying on propagating to the right, but a portion of the plane waves scatter off the post with the result that a circular wave travels outwards from the post. The outgoing circular wave interferes with the plane waves resulting in an interference pattern.

Now, what about scattering in quantum mechanics? Many things are different. For classical waves:

* Classical waves represent the amplitude of an observable wave.
* The differential equation governing the wave is the wave equation.
* The motion of the water is observable and predictable.

For quantum scattering:

* Quantum waves represent the amplitude of the wave function.
* The differential equation governing the wave function is the Schrodinger equation.
* The modulus squared of the wave function gives the probability that the WIMP will be observed at that point.
* The wave function representing the final state is predictable.
* However, this does not mean that the WIMP's final trajectory is predictable, because the wave function
only gives a probability distribution for where the WIMP goes after scattering.

What about the process of calculating the wave function?

* The wave function cannot be precisely calculated as Schrodinger's equation cannot be solved exactly.
* However, we can solve approximately for the final state of the WIMP using perturbation theory.

## Perturbation theory analysis of WIMP nuclear scattering

Start by picking a quantum mechanical problem for which there is a known exact solution. In non-relativistic quantum mechanics there are four:

* The infinite square well,
* The harmonic oscillator,
* The free particle,
* Central potentials (one-electron atoms).

Pick the solvable problem that most closely resembles the physical system you are analysing and write down the wavefunctions of this exactly soluble problem that will represent your incoming and outgoing states. In our case, the WIMP is essentially free both a long time before it interacts with the nucleus and a long time after. So, we pick the free particle as the closest exactly soluble problem to the one we are trying to analyse. 

### Plane wave to represent the incoming WIMP

The thing that we know about the incoming WIMP is its momentum, $\vec{p_w}$. We used this symbol in Lecture 2 to represent the momentum of the incoming WIMP. The wave function for a plane wave travelling in the direction of increasing $x$ is

$$\begin{align*}
\Psi_i(t,\vec{r})=Ae^{\frac{i}{\hbar}\left(\vec{p_w}\cdot\vec{r} - E_i t\right)},
\end{align*}$$

where $\vec{r}=x\hat{i}+y\hat{j}+z\hat{k}$,
and $E$ is the total non-relativistic energy of the particle.
Because we are not considering any interaction potential, the total energy is the kinetic energy,
$E=p_w^2/2m_w$, where $m_w$ is the WIMP mass. The energy stored in the rest mass of the WIMP plays no part
in this problem, because no mass energy gets converted into energy in the non-relativistic limit. 

### Plane wave to represent the outgoing WIMP

The outgoing
WIMP is also modelled as a plane wave, since without the interaction there is no significance to the origin. This isn't quite like the plane waves on water scattering on a post, because that problem and its physical picture of scattered circular waves represent an exact solution to the wave equation including the post. Here we are making our initial and final states as if the nucleus was not there at all. So, the final state
of the scattered WIMP is represented by

$$\begin{align*}
\Psi_f(t,\vec{r})=Ae^{\frac{i}{\hbar}\left(\vec{p_f}\cdot\vec{r} - E_f t\right)}.
\end{align*}$$

The figure below illustrates the initial and final plane waves represented by these wave functions.

### First order perturbation theory of the interaction

Now we've got some wave functions to represent the incoming and outgoing WIMP, we make the approximation that
defines first order perturbation theory.
The presence of the scattering potential does not alter the wave functions representing the initial
and final states. They take the same mathematical form that they would if the scattering nucleus was absent.

If you think about it you'll realise that ultimately this can't be exactly true, because
the potential $V(\vec{r})$ that describes the scattering potential of the nucleus will modify the Hamiltonian,
which defines the Schrodinger equation that must be solved to obtain the exact wave functions. 


However, in first order perturbation theory you assume you can keep the same wave functions. The entire effect of the scattering potential is to give you an amplitude for the particle initially in state $\Psi_i(t,\vec{r})$ to evolve into the same particle in state $\Psi_f(t,\vec{r})$. The scattering amplitude is given by an overlap integral between the initial and final states, with the time independent potential $V(\vec{r})$ representing the
field of the scattering nucleus connecting these two states.

$$\begin{align*}
a_{fi}&=\int\,dt \int\,d^3\vec{r}\,\Psi^*_f(t,\vec{r})V(\vec{r})\Psi_i(t,\vec{r}). \\
&=\int\,dt \int\,d^3\vec{r}\,e^{-\frac{i}{\hbar}(\vec{p_f}\cdot{r}-E_f t)}V(\vec{r})
e^{+\frac{i}{\hbar}(\vec{p_w}\cdot{r}-E_i t)}. \\
&=\int\,dt \int\,d^3\vec{r}\,e^{\frac{+iE_ft}{\hbar}}e^{\frac{-i\vec{p_f}\cdot\vec{r}}{\hbar}}
V(\vec{r})e^{\frac{-iE_it}{\hbar}}e^{\frac{+i\vec{p_w}\cdot\vec{r}}{\hbar}}
\end{align*}$$

Now, notice that the exponentials containing time $t$ don't have any dependence on $\vec{r}$, so they can
be brought outside the spatial integral,

$$\begin{align*}
a_{fi}&=\int\,dt\,e^{\frac{+iE_ft}{\hbar}}e^{\frac{-iE_it}{\hbar}}
 \int\,d^3\vec{r}\,e^{\frac{-i\vec{p_f}\cdot\vec{r}}{\hbar}}e^{\frac{+i\vec{p_w}\cdot\vec{r}}{\hbar}} \\
 &=\int\,dt\,e^{\frac{+i(E_f-E_i)t}{\hbar}}
\int\,d^3\vec{r}\,e^{\frac{-i(\vec{p_f}-\vec{p_i})\cdot\vec{r}}{\hbar}}
\tag{3.1}
\end{align*}$$

### The integral over time

The left hand integral over time is in fact a form of the Dirac $\delta$ that you may remember from David Mowbray's lectures in semester 1 of year 2. I know he teaches it because I've been a second year tutor for two decades! For those of you who are joining us for year 4, the result is

$$\begin{align*}
\int\,dx\,e^{i(k_1-k_2)x}\,dx&=2\pi\delta(k_1-k_2).
\end{align*}$$

Therefore, it follows that

$$\begin{align*}
\int\,dt\,e^{\frac{+i(E_f-E_i)t}{\hbar}}&=2\pi\delta\left(\frac{E_f}{\hbar}-\frac{E_i}{\hbar}\right) \\
&=\frac{2\pi}{\hbar}\delta(E_f-E_i)
\tag{3.2}
\end{align*}$$

This integral can therefore be interpreted as enforcing energy conservation. In a real scattering event, as we have seen from the hard sphere model, the WIMP recoils, and hence energy conservation leads to the sum of the nuclear recoil energy and the wimp recoil energy equaling the incident WIMP energy. In this quantum mechanical analysis, we are approximating the nucleus as a stationary scattering potential, because we have explicitly written $V(\vec{r})$, so the potential doesn't have any time dependence. It is not surprising that this leads to a requirement that the initial and final energies of the WIMP are the same. It's an approximate treatment, but as you will see it does allow us to learn some other interesting things about the interaction - interesting but not particularly fortunate if you are an experimentalist.

### The integral over space

The spatial integral is

$$\begin{align*}
\tilde{V}_{fi}=\int\,d^3\vec{r}\,V(\vec{r})e^{\frac{-i(\vec{p_f}-\vec{p_w})\cdot\vec{r}}{\hbar}}
\end{align*}$$

This integral is called the form factor of the nucleus.
Entire physicist's careers have been spent calculating and measuring nuclear form factors, and we
do not have space to do this physics-cum-art-form justice here. However, we can learn something from a 
toy model that is the next best thing to a real nuclear potential. First, we write $\vec{q}=\vec{p_w}-\vec{p_f}$.
The variable $\vec{q}$ is called the momentum transfer, because by conservation of momentum the difference
between the initial and final momenta of the wimp is the momentum transferred to the nucleus. The form
factor then becomes

$$\begin{align*}
F(\vec{q})=\int\,d^3\vec{r}\,V(\vec{r})e^{\frac{i\vec{q}\cdot\vec{r}}{\hbar}}
\end{align*}$$

Written this way, $F(\vec{q})$ has the form, up to constants, of the Fourier transform of the 
nuclear potential $V(\vec{r})$. This is analogous to the diffraction pattern of an optical beam
passing through an aperture being the Fourier transform of the aperture, something else that David 
Mowbray taught you in year 2, if you were here then.

In order to do this integral we need to assume a form for $V(\vec{r})$, the scattering potential
of the nucleus. A simple and physically-motivated choice is to stick close to the hard sphere scattering
model we started with, and put the interaction anywhere on the surface of a sphere of radius $R$. Therefore
we write $V(\vec{r})=V_0\delta(r-R)$, where $R$ is the radius of the nucleus. We align the momentum transfer
vector $\vec{q}$ with the $z$ axis and put the integral in spherical polar coordinates. We are free to 
do this as the integral is spherically symmetric apart from the dot product $\vec{q}\cdot\vec{r}$, and
with this choice of coordinates $\vec{q}\cdot\vec{r}=qr\cos\theta$, a fact that will allow us to do the
integral more easily. The figure below shows the chosen geometry and defines the angle $\theta$ and the 
potential $V(r)$.

![form_factor_integral](./figures/form_factor_integral.png)

In spherical coordinates, $(r,\theta,\phi)$, not forgetting the Jacobian factor in the 
change of variables so that $dx\,dy\,dz=r^2\sin\theta\,dr\,d\theta\,d\phi$, the form factor
integral becomes

$$\begin{align*}
F(\vec{q})&=\int_{\phi=0}^{2\pi}\,d\phi\int_{r=0}^\infty r^2V_0\delta(r-R)\,dr\int_{\theta=0}^\pi \sin\theta 
e^{\frac{iqr\cos\theta}{\hbar}}\,d\theta.
\end{align*}$$

The integral over $\theta$ is done first. If we let $u=\cos\theta/\hbar$ then $\hbar du=-\sin\theta\,d\theta$.
Substituting in this change of variables, and noting that the minus sign is absorbed by interchanging the top
and bottom limits, we obtain

$$\begin{align*}
F(\vec{q})&=\int_{\phi=0}^{2\pi}\,d\phi\int_{r=0}^\infty 
r^2V_0\delta(r-R)\,dr\,\hbar\int_{u=-1}^{+1} e^{iqRu}\,du \\
&=\int_{\phi=0}^{2\pi}\,d\phi\int_{r=0}^\infty 
r^2V_0\delta(r-R)\,dr\,\hbar\left[\frac{e^{iqru}}{iqr}\right]_{-1}^{+1} \\
&=\int_{\phi=0}^{2\pi}\,d\phi\int_{r=0}^\infty 
r^2V_0\delta(r-R)\,dr\,\frac{2\hbar}{qr}\left(\frac{e^{iqr}-e^{-iqr}}{2i}\right) \\
&=\int_{\phi=0}^{2\pi}\,d\phi\int_{r=0}^\infty 
r^2V_0\delta(r-R)\,dr\,\frac{2\hbar\sin{qr}}{qr} \\
&=4\pi R^3 V_0\hbar\frac{\sin{qR}}{qR},
\tag{3.3}
\end{align*}$$

where for the final step the Delta $\delta(r-R)$ eliminates the integral over $r$, where $r$ is equal to
the nuclear radius $R$. Physically, in our model, scattering of the WIMP of the surface of the nucleus
dominates.

Equations 3.2 and 3.3 then get substituted into Equation 3.1 so that $a_{fi}$, the scattering amplitude to first order in perturbation theory, is given by

$$\begin{align*}
a_{fi}=\frac{2\pi}\delta(E_f-E_i)4\pi R^3 V_0\frac{\sin{qR}}{qR}
\end{align*}$$

The probability of the scattering is proportional to $|a_{fi}|^2$. This expression is not so easy to square exactly because it contains the square of a delta. The important point for the physics, however, is that the
$|a_{fi}|^2$ is proportional to $F^2(q)$, equal to

$$\begin{align*}
F^2(q)&= \frac{\sin^2(qR)}{(qR)^2}
\end{align*}$$

In terms of the recoil energy $E_R$ and nuclear mass $m_N$, $E_R=q^2/(2m_N)$ so that

$$\begin{align*}
F^2(E_R)&= \frac{\sin^2(R\sqrt{2m_NE_R})}{2m_nE_RR^2}
\end{align*}$$

This drops off rapidly with increasing $E_R$, so that the more momentum is transferred to the nucleus, the more the form factor suppresses the probability of the scattering. This is a big departure from the picture in the hard sphere scattering model, where the probability distribution for the recoil energy was flat up to the maximum value allowed by conservation of energy and momentum. Here we expect very small values of momentum transferred to the nucleus to dominate. Furthermore, the function $\sin^2(qR)/(qR)^2={\rm sinc}^2(qR)$ has minima at 
$qR=n\pi$. So, just as in a diffraction pattern of a plane wave off a slit, there will be zeros in the interference pattern, corresponding to recoil energies which the nucleus cannot acquire according to this model.

More sophisticated form factor calculations also exhibit these zeros. We can plot the form factor in our crude model, and show some form factor calculations from publications that exhibit similar features. All of this
leads us to the idea that to be sensitive to WIMPs via the detection of nuclear recoils, we had better make
our detectors as sensitive as we can to nuclei recoiling with low kinetic energies.
