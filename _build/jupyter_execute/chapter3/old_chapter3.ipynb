First, we put the entire system, the WIMP, the nucleus, everything, in a box. This has the effect of making the space of states for the incoming and outgoing WIMP finite. We start by writing down Schrodingers equation for the free WIMP, without the interaction with the nucleus. The eigenstates of the time independent Schrodinger equation available to the WIMP are $\phi_n(\vec{x})$ corresponding to energy eigenvalues $E_n$ so that the time independent Schrodinger equation is $H_0\phi_n(\vec{x})=E_n\phi_n(x)$. The wave functions are the allowed momentum states for two particles, the WIMP and the nucleus, in a box. In the figure below the WIMP is in blue, the nucleus is in green. As in our hard sphere model we take the WIMP to be incoming from the left. Note that this picture shouldn't be taken too seriously; in the quantum model the states corresponding to propagating particles are in fact momentum eigenstates that look more like plane waves than balls.

![before_after](./figures/before_after/before_after.pdf)

 In the lab frame, the state corresponding to the initial state has the nucleus always at rest, but the basis of states also includes those representing the particles after collision, so the basis also includes all the states of constant momentum of each of the two particles. So in the above figure, A and B represent states that could be the initial state in the lab before the collision. C, D, I, J and L represent possible final states. The rest of the boxes contain states that, though they are members of the set of states that are solutions of the unperturbed Hamiltonian, they would play no part in a physical elastic collision process, either as an initial state or a final state.

The states representing the WIMP and nuclear motion after the collision have both of them moving if a collision occurred. The different possible states in the box obey an orthonormality relation, so that

$$\begin{align}
\int\phi_m^*(\vec{x})\phi_n(\vec{x})\,d\vec{x}&=\delta_{mn}
\tag{3.1}
\end{align}$$

Now, following the usual method for perturbation theory, we modify the Hamiltonian to include the scattering nucleus. The new Hamiltonian is $H=H_0+V(\vec{x})$. The modified Schrodinger equation is

$$\begin{align}
(H_0+V(\vec{x}))\Psi(\vec{x},t)=i\hbar\frac{\partial\Psi(\vec{x},t)}{\partial t}.
\tag{3.2}
\end{align}$$

After the collision the state of the WIMP has changed to $\psi(x)$, but in perturbation theory this state is expressed as a superposition of the states of the unperturbed Hamiltonian. Each of these states is an energy eigenstate, so each component state time evolves as a phase factor proportional to its energy.

![state space](./figures/unperturbed_states/unperturbed_states.pdf)

$$\begin{align}
\Psi(\vec{x},t)&=\sum_n a_n(t)\phi_n(\vec{x})e^{-iE_n t/\hbar},
\tag{3.3}
\end{align}$$

where the $a_n$ are complex coefficients of the different states in the superposition which in general can be time dependent. We substitute this back into the Schrodinger equation \ref{2.5} and obtain

$$\begin{align}
&\sum_n a_n(t)H_0\phi_n(\vec{x})e^{-iE_nt/\hbar}
+\sum_n a_n(t)V(\vec{x})\psi_n(\vec{x})e^{-iE_nt/\hbar} \nonumber \\
&=i\hbar\sum_n\phi_n(\vec{x})e^{-iE_nt/\hbar}\frac{da_n}{dt}
+i\hbar\sum_n\phi_n(\vec{x})a_n(t)\frac{-iE_n}{\hbar}a_n(t)e^{-iE_nt/\hbar}.
\tag{3.4}
\end{align}$$

Because $H_0\phi_n(\vec{x})=E_n\phi_n(\vec{x})$, the first and last terms cancel, leaving us with

$$\begin{align}
\sum_n a_n(t)V(\vec{x})\psi_n(\vec{x})e^{-iE_nt/\hbar}
&=i\hbar\sum_n\phi_n(\vec{x})e^{-iE_nt/\hbar}\frac{da_n}{dt}.
\tag{3.5}
\end{align}$$

The next step is to multiply by $\phi_m^*(\vec{x})$ and integrate over all space.

$$\begin{align}
\sum_n a_n(t)\int\phi_m^*(\vec{x})V(\vec{x})\phi_n(\vec{x})d^3x\,e^{-iE_nt/\hbar}
&= i\hbar\sum_n\int\phi_m^*(\vec{x})\phi_n(\vec{x})d^3x\,e^{-iE_nt/\hbar}\frac{da_n}{dt}
\nonumber \\
&= i\hbar\sum_n \delta_{mn}e^{-iE_nt/\hbar}\frac{da_n}{dt} \nonumber \\
&= i\hbar\frac{da_m}{dt}e^{-iE_mt/\hbar}.
\tag{3.6}
\end{align}$$

Rearranging this equation, we arrive at

$$\begin{align}
\frac{da_m}{dt}&=
\frac{-i}{\hbar}\sum_n a_n(t)e^{i(E_m-E_n)t/\hbar}
\int\phi_m^*(\vec{x})V(\vec{x})\phi_n(\vec{x})\,d^3x. 
\tag{3.7}
\end{align}$$

Next we consider that, to first order in perturbation theory, the time rate of change of $a_m$, the coefficient of the mth wavefunction of the unperturbed hamiltonian in the wavefunction of the scattered wimp, will be dominated by one particular input wave function. We therefore denote the output wave function by $m=f$, for final, and the input wavefunction we take to be dominated by one particular $n$, $n=i$, for initial. The dominance of one particular term allows us to dispence with the sum, and set $a_i=1$. We therefore get

$$\begin{align}
\frac{da_f}{dt}&=\frac{-i}{\hbar}
e^{i(E_f-E_i)t/\hbar}\int\phi_f^*(\vec{x})V(\vec{x})\phi_i(\vec{x})\,d^3x
\tag{3.8}
\end{align}$$

We interpret $da_f/dt$ as the rate at which the amplitude for the final state wave function is accumulating during the scattering, as time evolves. The integral over space represents the quantum mechanics of the interaction between the WIMP and the nucleus. We shall denote this integral by $V_{fi}$. It is often called the matrix element for the process. We shall see that it has a bearing on the scattering probability.

Our final piece of interpretation to interpret the integral of $da_f/dt$, which we call $a_{fi}$. It is a complex amplitude, and as usual in quantum mechanics its modulus squared is going to contribute to a probability. The interpretation is that it is proportional to the probability of the initial unscattered state $\phi_i$ evolving to the particular final state $\phi_f$ as a consequence of the action of the potential $V\vec{x}$.

$$\begin{align}
a_{fi}&=\frac{-i}{\hbar}V_{fi}\int_{t=-\infty}^{+\infty}\,dt\,e^{i(E_f-E_i)t/\hbar}
\tag{3.9}
\end{align}$$

We let $u=t/\hbar$ so that $dt=\hbar du$, and this integral becomes

$$\begin{align}
a_{fi}&=\frac{-i}{\hbar}V_{fi}\int_{u=-\infty}^{+\infty}\,du\,e^{i(E_f-E_i)u}
\tag{3.10}
\end{align}$$

The integral is an example of a Dirac delta that you were shown in second year when you studied Fourier transforms,

$$\begin{align}
\int_{u=-\infty}^{+\infty}\,du\,e^{i(E_f-E_i)u}&=2\pi\delta(E_f-E_i). 
\tag{3.11}
\end{align}$$

Applying this result we obtain an expression for $a_f$ after the interaction has taken place,

$$\begin{align}
a_{fi}=-2\pi i\delta(E_f-E_i)V_{fi}
\tag{3.12}
\end{align}$$

The appearance of the Dirac delta in the energy enforces overall energy conservation - the final state (including the recoiling WIMP) must have the same total energy as the initial state. Because this is elastic scattering, the sum of the masses before and after the collision is also conserved, hence the total kinetic energy is also conserved in the collision, just as it was in the classical hard sphere model.

It is not so easy to figure out the normalisation for this scattering amplitude. It is more fruitful to instead calculate $|a_{fi}|^2/T$, where $T$ is the interaction time. This quantity has the units of a rate, and is called the rate for the conversion of WIMPs into this particular final state.

We combine Equation 2.13 with the complex conjugate of Equation 2.15, to obtain

$$\begin{align}
R_{fi}=\frac{a_{fi}a_{fi}^*}{T}&=\frac{2\pi}{\hbar}\frac{V_{fi}V_{fi}^*}{T}\delta(E_f-E_i)
\int_{t=-T/2}^{+T/2}\,dt\,e^{i(E_f-E_i)t/\hbar} \nonumber \\
&=\frac{2\pi}{\hbar}\frac{|V_{fi}|^2}{T}\delta(E_f-E_i)\int_{t=-T/2}^{+T/2}\,dt \nonumber \\
&=\frac{2\pi}{\hbar}|V_{fi}|^2\delta(E_f-E_i). 
\tag{3.13}
\end{align}$$

The physical meaning of this result, known as the Fermi golden rule, is as follows. $R_{fi}$ is the rate of conversion of incoming WIMPs into this particular final state, which we take to be the WIMP and nucleus having undergone elastic scattering. The Dirac delta means that the process only has a non-zero rate if total energy, and hence kinetic energy, is conserved. If total energy is conserved, then the rate for this particular process is given by the modulus squared of the matrix element, $V_{fi}$. This matrix element contains the quantum physics of the interaction between the WIMP and the nucleus. Its detailed form is dependent on the exact physics behind the WIMP in question, but $V_{fi}$ may be diagrammatically represented as below.

![scattering vertex](./figures/scattering_vertex/scattering_vertex.pdf)

Evaluation of these matrix elements turns out to have quite an important impact on the rate of WIMP nuclear elastic scattering, and is the first part of the connection between the WIMP nuclear and WIMP nucleon elastic scattering cross sections.

The final important observation is that there may be more than one final state that has the same final state kinetic energy of the WIMP. We already know from the hard sphere scattering analysis that the final state kinetic energy of the WIMP is determined by the scattering angle $\theta$. However, the azimuthal angle $\phi$ can be anything, and still give rise to the same final state WIMP kinetic energy. If we are interested in the rate of a particular final state nuclear recoil energy, the rate is multiplied by the number of states in which the nuclear recoil energy takes a particular value. Classically, there are an infinite number of values of the azimuthal angle, but quantum mechanically the final state is quantised, and the number of possible states is, as alluded to in the hard sphere analysis, proportional to the magnitude of the final state momentum of the nucleus. This is closely analogous to the calculation of the density of the final states in statistical mechanics, only here one of the two angles is fixed, and only the other one is allowed to vary. So, the golden rule as sometimes written contains a factor usually denoted by $\rho(E_f)$, or $dN/dE_f$, customarily called the density of final states. This is in fact the degeneracy of final states that have the same final state observables. The density of final states turns out to be critical to the second part of working out the connection between the WIMP nuclear and WIMP nucleon elastic scattering cross sections.
