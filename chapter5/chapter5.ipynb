# Detectable Signals from Nuclear Recoils

## Light, and charge, and heat

The initial result of an elastic collision of a WIMP with a nucleus in a detector is a nucleus that has several $\rm keV$ of kinetic energy. This nucleus is moving through a detector containing other atoms. The binding energies of atomic electrons to their nuclei are a few $\rm eV$. Therefore, the kinetic energy of the recoiling nucleus is sufficient to cause many atoms to be excited. These excitations create signals of many kinds, and today we will talk only about the three most common. 

### Light and scintillation

Where the recoiling nucleus strikes an atom, it can cause one or more electrons in the struck atom to move from their ground state energy level to an excited state. The excited state electron then de-excites back into its ground state, and the energy lost results in the emission of a photon. This process is known as scintillation. Because the recoil energy of the nucleus is enough to cause multiple atoms to move to excited states, when these excited atoms de-excite, many photons may be emitted at once, resulting in a detectable light pulse. Some solids and liquids are known to be efficient scintillators. If the target, whether it is liquid or solid, is also transparent to the wavelength of light that is produced, then a sensitive light detector can be used to register the pulse. Either the height of the light pulse, or its area, can be used as a measure of the energy released in the pulse, and hence used to estimate the energy of the nuclear recoil.

### Charge and ionisation

The recoil energy is also enough to ionise many atoms, so that their electrons are freed from the bound states available to them entirely. These free electrons may recombine with the atoms that yielded them, or fall into another atom also ionised in the event, in which case again you see a scintillation light signal. However, if the detection medium is in a strong electric field, this field can draw the electron away from the ions and the charged electron can be transported through the detector, before being detected in some way with a sensor at the detector surface. The amount of ionisation charge in an event is again a measure of the recoil energy of the nucleus that caused the ionisation signal. Whether charge can be detected depends on whether free charges can move through the detector. Charge signals can be extracted from some solid and some liquid detectors.

### Heat

If the detector consists of a material having a crystalline structure, then the energy from a nuclear recoil can be converted into vibrationn energy in the crystal. The small amounts of energy involved mean that the excited heat entitites are phonons, which you may remember from solid state physics as the quanta of lattice vibrations. Phonons produce a very small amount of heating as they are collective excitations of the entire detector target, but the right sensing transducer is nonetheless capable of measuring the presence of phonons from the heating resulting from an elastic scattering event.

### Dual channel detection.

It is always better to design a detector so that it is sensitive to two different detectable signals, usually two of the ones described above. The reason is that the ratio of the energy deposited via these two different mechanisms is a powerful way of separating the nuclear recoils that are the signatures of WIMP scattering from the electron recoils produced by background non-WIMP events. More about backgrounds and background rejection later in the lecture. 

![light_charge_heat](./figures/threechannels.png)

There are essentially three different two-channel detector types. The first of these uses liquified noble gas, which both emits scintillation light, and allows for liberated charge to drift through the liquid. The second is crystals of semiconductor, germanium or silicon. These also allow charge transport. They are optically opaque so no scintillation can be detected, but when cooled to very low temperatures they can be instrumented with very sensitive thermometers that allow the detection of heating phonons. Finally, there are crystals made of calcium tungstate, which allow the detection of heating phonons and also scintillation light, but are electrically insulating so that electric charge cannot translate to the surface. 

Of the three detection technologies, the one that has succeeded in moving to the front, in terms of sensitivity, is noble liquid detector targets with the detection of scintillation light and charge. The main reason for this is that the noble liquid detectors are scalable to very large target masses, leading to higher event rates from a given WIMP population, as well as better self shielding, meaning that the target material itself acts effectively as a shield against background events reaching the heart of the detector.

## LUX-ZEPLIN - the most sensitive WIMP detector today.

Below is a diagram of the LUX-ZEPLIN experiment showing the dual signal readout of a large liquid xenon dark matter target via primary scintillation and the detection of ionisation electrons. The detector is about the height of a grown person.

![lux_zeplin](./figures/lux_zeplin.png)

LUX-ZEPLIN is the worlds leading WIMP detector. It has to date placed limits on WIMP dark matter with WIMP nucleon cross sections as low as $\rm 2\times10^{-12}\,pb$. The core target consists of liquid xenon, a noble group 8 element that liqifies at $\rm -110^\circ C$. This very cold tub of liquid xenon is over $\rm 1000\,kg$ in mass. 

#### Scintillation in LUX-ZEPLIN

The transparent xenon scintillates when a nucleus in the detector recoils, hopefully as a result of a collision with a dark matter WIMP. This scintillation light is emitted in all directions, and this scintillation light is detected in arrays of light sensors called photomultiplied tubes mounted on the top and bottom of the detector. The light emmitted in xenon scintillation is at a wavelength of around $\rm 178\,nm$, a vacuum ultra-violet wavelength. Therefore the scintillation light detectors have to be selected to have sensitivity at this wavelength. The light is detected with an array of photomultiplier tubes, a vacuum tube based light detector whose operation will be discussed later in the lecture. The choice of photomultiplier tubes is made because they have large collection area, reasonably high quantum efficiency, and manageable levels of noise. The signal resulting from the primary scintillation light generated close to the point where the original nuclear recoil occurred is referred to pretty universally as the S1 signal.

#### Ionisation in LUX-ZEPLIN

In addition, scintillation in xenon ionises some of the xenon atoms. The liberated negative electrons find themselves in an electric field that points downwards. The electric field is a result of a high voltage between a grid of fine wires mounted above the top of the xenon liquid, but below the photomultiplier tubes at the top, and another grid of fine wires mounted in the liquid above the bottom photomultiplier tubes. The field is kept uniform because surrounding the xenon cylinder are finely spaced metallic rings called field shaping rings, held at voltages becoming more positive linearly from the cathode electrode at the top to the anode electrode at the bottom. The electrons, being negative, start to accelerate towards the cathode, but as they are in a liquid they reach a terminal velocity that depends on the voltage, but is of order $\rm 1000\,m\,s^{-1}$. They continue at this terminal velocity until they reach the surface of the liquid. The detector is designed so that the surface of the liquid xenon is threaded by a larger electric field, in which the electrons break free of the liquid and move into the xenon gas above the lake. Once in the gas, the electrons accelerate rapidly, and this acceleration causes them to emit further light through the electroluminescence effect, resulting in a second signal in the photomultiplier tube array. This secondary electroluminescence signal is commonly referred to as the S2 signal.

#### Information from the signals in LUX-ZEPLIN

The S2 signal occurs after the S1 signal because it is delayed by the time it takes the electrons to drift through the xenon. Therefore the time delay between these two signals tells you the depth into the xenon where the event occurred. The direction of the electron drift is directly vertically upwards, and the S2 signal occurs very close to the photomultiplier tube faces. By measuring the ratio of the S2 signals produced in the different photomultipler tubes, an estimate of the two horizontal components of the event can be made. So, using the time delay between S1 and S2 and the horizontal reconstructed position from the sizes of the different photomultiplier signals, a full 3-dimensional reconstruction of the position where the event occurred can be carried out. This will turn out to be crucial for the rejection of background noise in the detector.

#### Principle of operation of photomultiplier tubes

A photomultiplier tube consists of a photocathode, a material that emits an electron when light falls on it. Typically the photocathode is a layer of material deposited on the glass or quartz envelope in the input end of the photomultiplier tube. Caesium iodide is commonly used as it has a suitable work function for the liberation of electrons by incoming ultraviolet light.  The photocathode is also maintained at a large negative voltage, relative to a series of so-called dynodes arranged in a ladder-like configuration as shown in the diagram below. The entire arrangement of photocathode and electrode chain is under high vacuum.

![pmt_operation](./figures/pmt.png)

The electron liberated from the photocathode accelerates in the electric field resulting from the voltage between the photocathode and the first dynode. In the vacuum, this electron accelerates in the electric field so that by the time it strikes the first dynode it has enough kinetic energy to release several electrons from the dynode surface. These electrons accelerate towards the second dynode, where each of them releases several secondary electrons. Many photomultipliers have as many as 14 dynodes in the chain. The succession of accelerations between dynodes leads to an avalanche in the population of electrons travelling down the chain, so that by the end of the chain there is a measurable charge pulse, which is then amplified.

Photomultiplier tubes typically have a quantum efficiency of 10s of percent, meaning that the probability of an individual incoming photon leading to an output voltage pulse is 10s of percent. This is quite a high efficiency for the detection of a single quantum of radiation. Photomultiplier tubes are also the source of two main varieties of background noise. First, the photomultiplier tube can produce output pulses when an electron is liberated from a dynode in the chain spontaneously. These are called noise pulses. They can mimic genuine pulses from scintillation light. Second, the photomultiplier tube is quite a complex assembly of elements. The outer vacuum envelope is made of some material, typically glass, but special low background photomultiplier tubes using metal as the vacuum envelope are commonly adopted by dark matter searches. The front window of the tube cannot be metal, and the preferred material for low background tubes is quartz. There then has to be a quartz to metal seal that joins the body to the front window, and this seal inevitably introduces other materials. It is possible that the photomultiplier tubes are themselves the source of a low but measurable level of radioactivity that may increase the noise level in the instrument. Though the background level of the PMTs can be made low by careful design, it is impossible to shield the detector from radiation produced by the photomultiplier tubes themselves as they have to be right next to the xenon target for the detector to be sensitive, so any residual radiation of the PMTs themselves presents a problem for the design of sensitive dark matter detectors.

Another issue with photomultiplier tubes is that they require a series of voltages on the dynodes to work. Traditionally a single high voltage supply was used to bias the cathode, and a chain of resistors connected successive dynodes to each other, so that the voltage on each successive dynode was less negative than the last. However, all these resistors and wires present a problem when there are hundreds of tubes, and some of them are immersed in the liquid xenon. In practice, it becomes more effective to wire all the dynodes up in parallel, with a single high voltage source supplying the same dynode in each photomultiplier tube in parallel. A complex set of electrodes and wiring is necessary to ensure that the set of photomultiplier tubes is biased correctly.

### Principal of operation of phonon detectors for WIMP searches

The detectors sensitive to phonons utilise superconducting transition edge thermomoeters, typically made of tungsten. In particular, this technology is adopted on the super CDMS experiment. The operating principle exploits the transition that tungston makes from normal-conducting to superconducting at around $\rm 100\,mK$ physical temperature. At this temperature, the electrical resistance of a thin wafer of tungston will be highly temperature-dependent. A constant voltage supply sets up a constant voltage across the wafer. This voltage will generate heating in a normal-conducting metal. The voltage is adjusted so that the thermal power dissipated in the wafer, $V^2/R$, reaches an equilibrium with heat conducted out of the wafer via an engineered thermal pathway, so that it reaches equilibrium at a temperature exactly at the transition. 

![phonon_detector](./figures/phonon_detector_2.png)

The bias is thermally self-stabilising. Suppose the wafer heats up. This causes the electrical resistance of the sample to rise, resulting in a decrease in $V^2/R$, lowering the electrical heating power supplied to the wafer, so that the temperature drops again. If the temperature drops, the resistance decreases, and $V^2/R$ is greater, so more heat is applied. Thus, a constant voltage across the sample having the right set point will stabilise the tungsten exactly at the transition temperature. WIMP-related phonons also try to heat up the sample, causing the same effect. By monitoring the current supplied to the sample at constant voltage, short time duration current pulses indicate that heat energy is being deposited in the tungsten. This is the mechanism by which phonons are detected.

The tungstan slices are bonded to the upper and lower surfaces of a germanium detector, and also double as electrodes for detecting charge deposited on the germanium crystal surfaces. As in noble liquid detectors, an electric field is set up across the germainium to facilitate the drifting of charge liberated after nuclear recoils to the surface of the detector. Also mirroring the noble liquid detectors, there is a danger of surface contamination of the germanium detectors resulting in surface events. Therefore, the charge collection and phonon electrodes are segmented across the face of the detector for full three dimensional event position reconstruction and fiducialisation.

## Backgrounds

The signal we are trying to detect is recoils of nuclei in the target material caused by scattering of WIMPs off the nuclei. Other things can mimic this signal, most significantly two classes of phenomena. 

First, other non-WIMP particles can collide with nuclei and cause them to recoil. The primary source of background nuclear recoils is neutrons entering the detector. Neutrons are highly penetrating, meaning that they can pass through quite a high thickness of solid material. This is mainly because they are electrically neutral and hence have minimal electromagnetic interaction with nuclei and electrons in the material they are passing through. Neutron backgrounds can originate from cosmic rays, or from radioactive decays either in the detector itself or from the environment surrounding the detector.

Second, other non-WIMP particles can collide not with the nuclei in the detector, but with the electrons. An energetic electron produced in the detector can result in secondary scintillation light and ionisation, and these signals can potentially mimic a nuclear recoil. Though as we shall discuss there is a strategy for using the two different detectable signals together to achieve discrimination between nuclear and electron recoils, there are a large number of electron recoils because they can originate from background gamma radiaton in the detector. Gamma ray backgrounds are the most common source of background hits in WIMP detectors.

## Shielding

Cosmic rays are a major source of background spurious nuclear recoils. In a common mechanism, a cosmic ray muon scatters off a neutron in the vicinity of the detector, and the neutron enters the target and causes a nuclear recoil. Many more complex mechanisms and cascades originating in cosmic ray showers are possible. The primary defence against cosmic ray backgrounds is to operate the WIMP detector deep underground, and use the rock overburden above the detector to reduce the cosmic ray flux entering the detector. This is the main reason for the operation of deep underground labs for dark matter detection.

The radiation background at the detector also contains a component that originates from the rock surrounding the detector in its cavern. This leads to most WIMP detectors being surrounded by shielding designed to reduce the flux of radiation entering the detector volume. Smaller detectors commonly use lead shielding against gamma rays and hydrocarbon shielding (usually borated wax) to shield against neutrons. Large detectors like LUX-ZEPLIN use a second skin layer of liquid xenon as a shield, and also use a very large volume of water in a tank around the detector to shield against neutrons. 

The last strategy is to install a secondary active detector outside the primary detection volume. This is called an active shield. Any event coincident with an event detected in the active veto is discarded, as a real WIMP would interact in the detector only once. The LUX-ZEPLIN detector has an active veto made of gadolinium loaded liquid scintillator. Gadolinium has a very high neutron capture cross section for very low energy neutrons. The excited gadolinium then de-excites with emission of a gamma, leading to scintillation light that is picked up by more photomultiplier tubes. 

Finally the entire veto and detector is embedded in a water tank, the water being an effective shield against neutrons, as it contains many hydrogen atoms in the water molecules, having masses equal to that of the neutron. Neutrons are therefore effectively moderated in the water tank, hopefully not managing to penetrate to the veto or detector inside.

### Sources of radioactive 

There are many sources of background, one being cosmic rays. This is the reason why all sensitive detectors for WIMPS are run in underground laboratories. The most critical threat is cosmic ray muons scattering on or near the detector and causing ejection of neutrons that then contribute a nuclear recoil background. Underground operation at several kilometres water-equivalent in a laboratory such as Boulby reduces the cosmic ray flux by several orders of magnitude. Backgrounds are also produced by components of the detector itself, which may contain trace amounts of uranium and thorium that then result in the occasional decay following the chains discussed above. Also a problem is radon, produced in the Earth and filtering into the underground lab through the walls. As a gas, Radon can find its way into the Xenon itself, and a strict regimen of chemical cleaning of the Xenon and recirculation of Xenon in the detector through a cleaning apparatus is a key part of keeping backgrounds in the detector minimal.

### Fiducialisation

A final defence against backgrounds is to reconstruct where events took place in the detector and discard events whose reconstructed position is close to the detector surface. It turns out that the majority of the hits in LUX-ZEPLIN. The figure below shows the reconstructed positions of events in LUX-ZEPLIN. The vast majority of events are close to the side walls, the detector bottom, and the top electrode. These events are eliminated as canidates by discarding all events outside an acceptance region defined by the black line. The smaller this acceptance region, the smaller the effective mass of the detector, so there is trade-off between fiducialisation veto efficiency and detector mass.

### Limit

As of January 2024, LUX-ZEPLIN succeeded in placing limits on the cross section for WIMP nucleon elastic scattering as low as $\rm 2\times10^{-12}\,pb$. To put this in context, this is roughly five orders of magnitude more stringent than the best results from when I left the WIMP dark matter collaborations in 2008. It is a remarkable piece of work, of which the collaborators in this endeavour should be very proud.
