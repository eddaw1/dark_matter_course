# Dark matter candidates and detectors

As discussed in the preceding lectures by Vik Dhillon, either we modify our ideas of classial mechanics in the low acceleration limit, or we modify the size and matter contents of galaxies and clusters to include a large amount of extra, unidentified dark matter beyond the baryonic matter we already observe. In these lectures I will focus on the latter possibility, that there exists a large amount of undiscovered dark matter. I discuss the main classes of dark matter candidates and the instruments that are operating, and planned, to try and detect these dark matter candidates. 

## Dark matter might have several or many constituents

Importantly, the existence of dark matter is so far inferred only from its gravitational interactions. Gravity couples to all objects that have mass. The total mass of a dark matter cloud is the sum of the masses of its constituents, whether those constituents are the same or different from each other. Therefore, any combination of a variety of dark matter types could act collectively as the dark matter, resulting in the gravitational effects we observe. 

For example, we already know that some MACHO dark matter is out there, but it is insufficiently abundant to explain the observed gravitational effects on its own. The remaining dark matter could be WIMPs, axions, dwarf black holes, other anticipated and/or unanticipated stable massive particle(s), or any combination of the above! The possibility, or even likelihood, of mixed dark matter, would be bad news for experimentalists. The already faint signals predicted to arise from any one of the candidates would be commensurately fainter. Nature does not, however, care how difficult things are for us in the laboratory.
